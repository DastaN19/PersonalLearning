```python
# importing important libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# import tensorflow libraries
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
ls
```

```python
# load the dataset
data = pd.read_csv('Churn_Modelling.csv')
data
```

```python
# segregate the independetn and dependetn varibale
X = data.iloc[:,3:13]
y = data.iloc[:,13]
```

```python
X
```

```python
# checing for missing values

X.isnull().sum()
```

```python
geography = pd.get_dummies(X['Geography'], drop_first=True)
gender = pd.get_dummies(X['Gender'], drop_first=True)
```

```python
X = pd.concat([X,geography,gender], axis = 1)
```

```python
X
```

```python
# dropping the unwanted columns
X = X.drop(columns=['Geography','Gender'], axis=1)
```

```python
X.head()
```

```python
# splitiing the datase

X_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(x_test)
```

```python
# creation of model

model = Sequential()

model.add(Dense(200,activation='sigmoid',kernel_initializer='glorot_uniform')) # hidden layer 1
model.add(Dropout(0.2))

model.add(Dense(100, activation='relu', kernel_initializer='he_uniform')) # hidden layer 2
model.add(Dropout(0.4))

model.add(Dense(50, activation = 'sigmoid', kernel_initializer='glorot_uniform')) #hidden layer 3
model.add(Dropout(0.5))

model.add(Dense(10, activation = 'relu', kernel_initializer='he_uniform')) # hidden layer 4

model.add(Dense(1)) #output layer
```

```python
# compiling the model

model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size = 20, validation_data = (X_test,y_test))
```

```python
model.evaluate(X_test , y_test)
```
