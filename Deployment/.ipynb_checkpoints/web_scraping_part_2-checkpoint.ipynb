```python
import requests
import os
from bs4 import BeautifulSoup as bs
```

```python
def get_html(from_year,till_year):
    for year in range(from_year , till_year+1):
        for month in range(1,13):
            if month < 10:
                url = f'https://en.tutiempo.net/climate/0{month}-{year}/ws-421820.html'
            else:
                url = f'https://en.tutiempo.net/climate/{month}-{year}/ws-421820.html'
            
            data = requests.get(url).text
        
            if not os.path.exists(f"data/html_data/{year}"):
                os.makedirs(f"data/html_data/{year}")
            
        
            with open(f"data/html_data/{year}/{month}.html",'w') as f:
                f.write(data)
```

```python
get_html(2013,2019)
```

```python
def get_csv(month,year):
    
    file_html = open(f"data/html_data/{year}/{month}.html" , 'r')
    plain_text = file_html.read()
    
    temp_d = []
    final_d = []
    
    soup = bs(plain_text)
    for table in soup.find('table',{'class':'medias mensuales numspan'}):
        for tbody in table:
            for tr in tbody:
                temp_d.append(tr.text)
                
    rows = len(temp_d)/15 #table has 15 columns
    for _ in range(round(rows)):
        new_temp_d = []
        for i in range(15):
            new_temp_d.append(temp_d[0])
            temp_d.pop(0)
        final_d.append(new_temp_d)
    return final_d
```

```python
import pandas as pd
df = pd.DataFrame(get_csv(1,2014))
df
```
